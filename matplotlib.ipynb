`generating sin and cos graph using matplotlib`

```python
x=np.linspace(0,5*np.pi,100)

#defining the wave
sin_y=np.sin(x)
cos_y=np.cos(x)

#plotting the wave
plt.plot(x,sin_y,'*')
plt.plot(x,cos_y, '-')
plt.legend(loc='upper right')
plt.title('SINE and COSINE Wave')
plt.show()
```

`legend is used to add labels to the graph given in the plot. different values for loc such as lower right , center , upper right, lower left etc.`
`title function adds a title to the plot`

`'-' will construct a line like in above plots`

#### DIFFERENT TYPES OF PLOTS 
* line plot
Used to visualise relationship between two continuous variables.usually used to study time series data , where the x axis represents time, and the y axis represents measurements.




```python
x=np.linspace(0,4*np.pi,100)
y=np.sin(x)
plt.style.use('Solarize_Light2')
plt.figure(figsize=(10,5))
plt.plot(x,y,label='sine')
plt.legend(loc='upper right')
plt.title('SINE and COSINE Wave')
plt.show()
```

```python
#multiple line plots
x=np.linspace(0,4*np.pi,100)
y1=np.sin(x)
y2=np.cos(x)
plt.style.use('Solarize_Light2')
plt.figure(figsize=(10,5))
plt.plot(x,y1,'*',label='sine',color='red',linewidth=1)
plt.plot(x,y2,'--',label='cos',color='blue',linewidth=1)
plt.legend(loc='upper right')
plt.title('SINE and COSINE Wave')
plt.show()
```

#### adjusting the plot axes
* can be done using plt.xlim for x axis
* plt.ylim for y axis

```python
x=np.linspace(0,4*np.pi,100)


plt.xlim([0,15])
plt.ylim([-2,2])
y=np.sin(x)


plt.plot(x,y,label='sine')
plt.legend(loc='upper right')
plt.title('SINE and COSINE Wave')
plt.show()
```

#### using .set to combine all of these commands

```python
x=np.linspace(0,4*np.pi,100)

fig=plt.axes()
fig.plot(x,np.sin(x))
fig.set(xlim=(0,10),ylim=(-2.5,2.5),xlabel='xaxis',ylabel='sin(x)',title='sine wave')
plt.legend(loc='upper right')
plt.title('SINE and COSINE Wave')
plt.show()
```

#### SCATTER PLOTS
* type of data visualisation in which individual data points are represented das points in a two dimensional plane. x coordinate represents one type of data and y coordinate represents another feature.


```python
n=500

plt.style.use('bmh')
x1=np.random.randn(n)
y1=np.random.randn(n)
x2=np.random.randn(n)+1
y2=np.random.randn(n)+1
plt.scatter(x1,y1,c='red',alpha=0.5,s=10,label='group1')
plt.scatter(x2,y2,c='blue',alpha=0.5,s=10,label='group2')
plt.legend(loc='upper right')
plt.title('scatter plot with multiple groups')
plt.show()
```

#### the main difference between plt.plot and plt.scatter is that plt.scatter gives us the freedom to control the properties of each individual point.(size,face_colour,edge_colour,transparency level etc)

```python
rng=np.random.RandomState(0)
x=rng.randn(100)
y=rng.randn(100)
colors=rng.rand(100)
sizes=rng.rand(100)*1000
plt.scatter(x,y,c=colors,s=sizes,alpha=1,cmap='viridis',edgecolors='black')
plt.title('random scatter plot')
plt.xlabel('x values')
plt.ylabel('y values')
plt.colorbar()
plt.show()
```

* alpha: transparency of the markers in the scatter plot
* colorbar: a bar showing the colors used in the plot and the corresponding values 
* colormap: convert scalar data to colors
* edge colors: the color of the edge of the markers in a scatter plot
* sizes: the sizes of the markers in a scatter plot.


#### HISTOGRAM AND BINNINGS

A histogram is a plot that displays the distribution of a set of continuous or discrete data.The data is divided into a set of intervals or bins. 
* binnings refers to the process of dividing the data into these bins

```python
data =np.random.randn(1000000)
plt.hist(data,bins=10,color='purple',edgecolor='purple',alpha=1,histtype='stepfilled')
plt.title('random histo')
plt.xlabel('values')
plt.ylabel('frequency')
plt.show()
```

#### Multiple Histograms


```python
data =np.random.randn(1000000)
data2 =np.random.randn(1000000)+2
data1 =np.random.randn(1000000)+4
plt.hist(data,bins=20,alpha=0.8,edgecolor='black')
plt.hist(data1,bins=20,alpha=0.8,edgecolor='black')
plt.hist(data2,bins=20,alpha=0.8,edgecolor='black')
plt.title('random histo')
plt.xlabel('values')
plt.ylabel('frequency')
plt.show()
```

#### 2D HISTOGRAM
plt.hist2d is fucntion in matplotlib that generates a 2D histogram.A 2D histogram is representation of the distribution of data over two variables. In contrast to a regular histogram which shows the distribution of a single variable, 2D histogram shows the relationship between two variables. The function creates a heatmap that shows the density of the data points in a 2D space.

```python
data2 =np.random.randn(1000)+2
data1 =np.random.randn(1000)+4
plt.hist2d(data1,data2,bins=20,cmap='viridis')
plt.title('2d histogram')
plt.xlabel('data1')
plt.ylabel('data2')
plt.show()
```

```python
data2 =np.random.randn(1000)+2
data1 =np.random.randn(1000)+4
plt.hist2d(data1,data2,bins=20,cmap='Blues')
plt.title('2d histogram')
plt.colorbar()
plt.xlabel('data1')
plt.ylabel('data2')
plt.show()
```

#### PIE CHARTS
A pie chart is a circular statistical graphic, which is divided into slices to illustrate numerical proportion.

`adding shadows`

```python
cars=["audi",'bmw','ford','tesla','jaguar','mercedes']
data=[450,345,35,100,200,500]
explode=[0.1,0.2,0.3,0.2,0.1,0.3]
fig=plt.figure(figsize=(7,7))
plt.pie(data,labels=cars,explode=explode,shadow=True)
plt.show()
```

`adding percentage`

```python
cars=["audi",'bmw','ford','tesla','jaguar','mercedes']
data=[450,345,35,100,200,500]
explode=[0.1,0.2,0.3,0.2,0.1,0.3]
fig=plt.figure(figsize=(7,7))
plt.pie(data,labels=cars,explode=explode,shadow=True,autopct='%2.1f%%')
plt.show()
```

 `multiple wedget`

```python
cars=["audi",'bmw','ford','tesla','jaguar','mercedes']
data=[450,345,35,100,200,500]
explode=[0.1,0.2,0.3,0.2,0.1,0.3]
fig=plt.figure(figsize=(7,7))
plt.pie(data,labels=cars,explode=explode,shadow=True,autopct='%2.1f%%',wedgeprops={'edgecolor':'black','linewidth':2,'antialiased':False})
plt.show()
```

#### Multiple Subplots 
multiple subplots refer to dividing a figure into multiple subplots or axes, each of which can contain a separate plot. This allows for multiple plots to be displayed in a single figure. In Matplotlib, subplots function can be used to for subplot.

```python
x=np.linspace(0,10,50)
fig,ax=plt.subplots(2,2)

y=np.sin(x)
ax[0,0].plot(x,y)
ax[0,0].set_title('Sin Wave')

y=np.cos(x)
ax[0,1].plot(x,y)
ax[0,1].set_title('cos Wave')

y=np.tan(x)
ax[1,0].plot(x,y)
ax[1,0].set_title('tan Wave')

y=np.random.rand(1000)
ax[1,1].hist(y,bins=40)
ax[1,1].set_title('Random Distribution')
plt.tight_layout()

plt.show()
```

```python
# with loops example 1
plt.subplots_adjust(hspace=0.4,wspace=0.4)
for i in range(1,7):
    plt.subplot(2,3,i)
    plt.text(0.5,0.5,str((2,3,i)),fontsize=18,ha='center')
```

```python
x=np.linspace(0,10,50)
y=np.sin(x)

fig,(ax1,ax2,ax3)=plt.subplots(1,3,figsize=(6,6))
ax1.plot(x,y)
ax2.plot(x,y**2)
ax3.plot(x,y**3)
ax1.set_title('sin wave')
ax2.set_title('sin wave square')
ax3.set_title('sin wave cubic' )
plt.tight_layout()
plt.show()
```
