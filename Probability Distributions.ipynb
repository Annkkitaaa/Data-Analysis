```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Parameters for the uniform distribution
a = 0  # Lower bound
b = 10  # Upper bound

# Create an array of x values
x = np.linspace(a - 1, b + 1, 1000)

# Calculate the PDF values for the uniform distribution
pdf_values = uniform.pdf(x, loc=a, scale=b - a)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, pdf_values, label='PDF: Uniform Distribution')

plt.title('Probability Density Function (PDF) of Uniform Distribution')
plt.xlabel('x')
plt.ylabel('PDF Value')
plt.legend()
plt.grid()

plt.show()
```

##### 5)Normal Distribution (also known as gaussian distribution)
most of the data basically cluster around average, with fewer and fewer value as one moves away from average in either direction.

EXAMPLE: 
* to analyze the height of people for clothing store to know which size of clothes require more






```python
from scipy.stats import norm
x=norm.rvs(0,1)
x
```

```python
x_data =np.arange(-5,5,0.0001)
y_data=norm.pdf(x_data,0,1)
plt.plot(x_data,y_data)
plt.show()
```

```python
z_data=norm.pdf(x_data,2,1)
plt.plot(x_data,z_data)
plt.show()
```
